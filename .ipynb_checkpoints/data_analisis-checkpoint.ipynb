```python
df_score = df[['Movie Title', 'Year of Realease', 'Movie Rating', 'Metascore', 'Votes', 'Worldwide LT Gross']].copy()

df_score['Worldwide LT Gross'] = df_score['Worldwide LT Gross'].replace({'\$': '', ',': ''}, regex=True)
df_score['Worldwide LT Gross'] = df_score['Worldwide LT Gross'].astype(float)
df_score['Votes'] = df_score['Votes'].replace(',', '', regex=True)
df_score['Votes'] = df_score['Votes'].astype(int)
df_score['Movie Rating'] = df_score['Movie Rating'].astype(float)
df_score['Metascore'] = df_score['Metascore'].replace('******', float('NaN'))
df_score['Metascore'] = df_score['Metascore'].astype(float)

df_score_top10 = df_score.sort_values(by=['Movie Rating', 'Votes'], ascending=False).head(10)

df_score_top10
```

```python
X_gross = df_gross['Gross'].values.reshape(-1,1)
y_gross = df_gross['Worldwide LT Gross'].values.reshape(-1,1)

X_gross_train, X_gross_test, y_gross_train, y_gross_test = train_test_split(X_gross, y_gross, test_size=0.2, random_state=0)

model_gross = LinearRegression()  
model_gross.fit(X_gross_train, y_gross_train)

y_gross_pred = model_gross.predict(X_gross_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_gross_test, y_gross_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_gross_test, y_gross_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_gross_test, y_gross_pred)))
```

```python
plt.scatter(X_gross_test, y_gross_test, color='gray')
plt.plot(X_gross_test, y_gross_pred, color='red', linewidth=2)
plt.show()
```

```python
alphas = [0.01, 0.1, 1.0, 10.0, 100.0]

ridge_gross = RidgeCV(alphas=alphas, cv=5)

ridge_gross.fit(X_gross_train, y_gross_train)

y_gross_pred = ridge_gross.predict(X_gross_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_gross_test, y_gross_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_gross_test, y_gross_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_gross_test, y_gross_pred)))
```

```python
y_gross_pred_orig = model_gross.predict(X_gross_test)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].scatter(y_gross_test, y_gross_pred_orig, color='blue')
ax[0].plot([y_gross_test.min(), y_gross_test.max()], [y_gross_test.min(), y_gross_test.max()], 'k--', lw=4)
ax[0].set_xlabel('Real')
ax[0].set_ylabel('Predicted')
ax[0].set_title('Real vs Predicted Values (Original Model)')

ax[1].scatter(y_gross_test, y_gross_pred, color='green')
ax[1].plot([y_gross_test.min(), y_gross_test.max()], [y_gross_test.min(), y_gross_test.max()], 'k--', lw=4)
ax[1].set_xlabel('Real')
ax[1].set_ylabel('Predicted')
ax[1].set_title('Real vs Predicted Values (Improved Model)')

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['Movie Title', 'Year of Realease', 'Genre']

column_trans = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), categorical_cols),
    remainder='passthrough'
)

linreg = LinearRegression()

pipe_genre = make_pipeline(column_trans, linreg)

X_genre = df_genre.drop('Worldwide LT Gross', axis=1)
y_genre = df_genre['Worldwide LT Gross']

X_genre_train, X_genre_test, y_genre_train, y_genre_test = train_test_split(X_genre, y_genre, test_size=0.2, random_state=0)

pipe_genre.fit(X_genre_train, y_genre_train)

y_genre_pred = pipe_genre.predict(X_genre_test)

mse = mean_squared_error(y_genre_test, y_genre_pred)
print('Mean Squared Error:', mse)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(y_genre_test, y_genre_pred, color='blue')
plt.plot([y_genre_test.min(), y_genre_test.max()], [y_genre_test.min(), y_genre_test.max()], 'k--', lw=2)
plt.xlabel('Real')
plt.ylabel('Predicted')
plt.title('Real vs Predicted Values')

plt.show()
```

```python
ridge_genre = Ridge(alpha=1.0)

pipe_genre = make_pipeline(column_trans, ridge_genre)

pipe_genre.fit(X_genre_train, y_genre_train)

y_genre_pred = pipe_genre.predict(X_genre_test)

mse = mean_squared_error(y_genre_test, y_genre_pred)
print('Mean Squared Error:', mse)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(y_genre_test, y_genre_pred, color='blue')
plt.plot([y_genre_test.min(), y_genre_test.max()], [y_genre_test.min(), y_genre_test.max()], 'k--', lw=2)
plt.xlabel('Real')
plt.ylabel('Predicted')
plt.title('Real vs Predicted Values for Ridge Regression')

plt.show()
```
